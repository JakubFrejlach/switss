```python
from switss.model import DTMC
import numpy as np
from switss.utils import color_from_hash

# https://en.wikipedia.org/wiki/Tarjan%27s_strongly_connected_components_algorithm#/media/File:Tarjan's_Algorithm_Animation.gif
# https://en.wikipedia.org/wiki/Tarjan%27s_strongly_connected_components_algorithm
# (same graph)
E = [[0,1],[1,2],[2,0],[3,2],[3,1],[3,5],[4,2],[4,6],[5,4],[5,3],[6,4],[7,5],[7,6],[7,7]]

# create transition matrix from graph (with arbitrary probabilities)
P = np.zeros(shape=(8,8))
for u,v in E:
    ucount = len([w for w,z in E if w == u])
    P[u,v] = 1/ucount

dtmc = DTMC(P)
components = dtmc.strongly_connected_components()

def sm(stateidx,labels):
    return { "color" : color_from_hash(4*components[stateidx]), "style" : "filled" } 

dtmc.digraph(state_map=sm)
```
